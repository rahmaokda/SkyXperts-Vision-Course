### Task 13: Implement Adaptive Median Filter
Write a custom function for adaptive median filtering. It should dynamically increase window size until noise is removed or max size is reached. Apply to a noisy image and compare with standard median.

```python
import cv2 as cv
import numpy as np
import sys
import matplotlib.pyplot as plt
%matplotlib inline

def add_salt_pepper_noise(image, density=0.05):
    # Implement using random pixels set to 0 or 255
    noisy_img = np.copy(image)
    num_salt_pepper = int(density * image.size)

    salt_row = np.random.randint(0, image.shape[0], int(num_salt_pepper/2))
    salt_col = np.random.randint(0, image.shape[1], int(num_salt_pepper/2))
    noisy_img[salt_row, salt_col] = 255

    pepper_row = np.random.randint(0, image.shape[0], int(num_salt_pepper/2))
    pepper_col = np.random.randint(0, image.shape[1], int(num_salt_pepper/2))
    noisy_img[pepper_row, pepper_col] = 0

    return noisy_img



# Your code here
def adaptive_median_filter(image, max_size=7):
    # Implement logic: for each pixel, start with small window, increase if needed
    padded_image = np.pad(image, max_size // 2, mode='reflect')
    output_image = np.copy(image)

    H, W, C = image.shape
    
    for i in range(H):
        for j in range(W):
            for k in range(C):
                
                s = 3 
                while s <= max_size:
                    half_s = s // 2
                
                    window = padded_image[i:i + s, j:j + s, k].flatten()
                
                    Z_min = np.min(window)
                    Z_max = np.max(window)
                    Z_med = np.median(window)
                    Z_xy = image[i, j, k]
                    
                    if Z_med > Z_min and Z_med < Z_max:
                        if Z_xy > Z_min and Z_xy < Z_max:
                            output_image[i, j, k] = Z_xy
                        else:
                            output_image[i, j, k] = Z_med
                        break 
                    else:
                        s += 2 
                        if s > max_size:
                            output_image[i, j, k] = Z_med

    return output_image


# Test on noisy image
path = './images/logo.png' 
img_bgr = cv.imread(path, 1)

img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

noise_density = 0.1
noisy_image = add_salt_pepper_noise(img_rgb, density=noise_density)

standard_median = cv.medianBlur(noisy_image, 5)

adaptive_median = adaptive_median_filter(noisy_image, max_size=5)

plt.figure(figsize=(15, 5))
titles = ['Noisy Image', 'Standard Median (5x5)', 'Adaptive Median (Max 5x5)']
images = [noisy_image, standard_median, adaptive_median]

for i, (image, title) in enumerate(zip(images, titles)):
    plt.subplot(1, 3, i + 1)
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

### Task 14: Implement Bilateral Filter Function
Write a Python function to perform bilateral filtering on an image. Use Gaussian weights for both spatial and intensity. Parameters: diameter, sigma_color, sigma_space. Compare with cv.bilateralFilter().

```python
import cv2 as cv
import numpy as np
import sys
import matplotlib.pyplot as plt


def add_salt_pepper_noise(image, density=0.05):
    # Implement using random pixels set to 0 or 255
    noisy_img = np.copy(image)
    num_salt_pepper = int(density * image.size)

    salt_row = np.random.randint(0, image.shape[0], int(num_salt_pepper/2))
    salt_col = np.random.randint(0, image.shape[1], int(num_salt_pepper/2))
    noisy_img[salt_row, salt_col] = 255

    pepper_row = np.random.randint(0, image.shape[0], int(num_salt_pepper/2))
    pepper_col = np.random.randint(0, image.shape[1], int(num_salt_pepper/2))
    noisy_img[pepper_row, pepper_col] = 0

    return noisy_img



# Your code here
def custom_bilateral_filter(image, diameter, sigma_color, sigma_space):
    # Implement using nested loops or vectorized (efficiently)
    # For each pixel, compute weighted sum based on distance and intensity diff
    output = np.zeros_like(image, dtype=np.float32) 
    radius = diameter // 2
    y, x = np.indices((diameter, diameter))
    center = (radius, radius)
    distance_sq = (x - center[0])**2 + (y - center[1])**2
    G_spatial = np.exp(-0.5 * (distance_sq / (sigma_space**2)))
    H, W, C = image.shape
    for c in range(C):
        img_c = image[:, :, c].astype(np.float32)
        
        for i in range(H):
            for j in range(W):
                I_p = img_c[i, j]
                i_start = max(0, i - radius)
                i_end = min(H, i + radius + 1)
                j_start = max(0, j - radius)
                j_end = min(W, j + radius + 1)
                window = img_c[i_start:i_end, j_start:j_end]
                k_i_start = i_start - (i - radius)
                k_i_end = diameter - ((i + radius + 1) - i_end)
                k_j_start = j_start - (j - radius)
                k_j_end = diameter - ((j + radius + 1) - j_end)
                
                kernel_window = G_spatial[k_i_start:k_i_end, k_j_start:k_j_end]
                diff_intensity_sq = (window - I_p)**2
                G_color = np.exp(-0.5 * (diff_intensity_sq / (sigma_color**2)))
   
                W_combined = kernel_window * G_color
         
                I_weighted_sum = np.sum(window * W_combined)
       
                W_sum = np.sum(W_combined)
        
                output[i, j, c] = I_weighted_sum / W_sum
    
    return output.astype(image.dtype) 


# Apply to image, display, and compare with OpenCV's version

path = './images/logo.png' 
img = cv.imread(path, 1)
noisy_image = add_salt_pepper_noise(img, density=0.003)

custom_filtered = custom_bilateral_filter(noisy_image, 9, 75, 75)

custom_filtered=cv.cvtColor(custom_filtered, cv.COLOR_BGR2RGB)

cv_filtered = cv.bilateralFilter(noisy_image, 9, 75, 75)

cv_filtered=cv.cvtColor(cv_filtered, cv.COLOR_BGR2RGB)

plt.figure(figsize=(15, 5))
titles = ['A. Noisy Input', 'B. Custom Bilateral Filter', 'C. cv.bilateralFilter() (OpenCV)']
images = [noisy_image, custom_filtered, cv_filtered]

for i, (image, title) in enumerate(zip(images, titles)):
    plt.subplot(1, 3, i + 1)
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

### [BONUS] Task 15: Comprehensive Camera Task 
Combine: Live camera feed -> grayscale -> add noise -> remove with median -> sharpen. Display all stages in separate windows.

```python
import cv2 as cv
import numpy as np
import sys

def add_salt_pepper_noise(image, density=0.05):
    # Implement using random pixels set to 0 or 255
    noisy_img = np.copy(image)
    num_salt_pepper = int(density * image.size)

    salt_row = np.random.randint(0, image.shape[0], int(num_salt_pepper/2))
    salt_col = np.random.randint(0, image.shape[1], int(num_salt_pepper/2))
    noisy_img[salt_row, salt_col] = 255

    pepper_row = np.random.randint(0, image.shape[0], int(num_salt_pepper/2))
    pepper_col = np.random.randint(0, image.shape[1], int(num_salt_pepper/2))
    noisy_img[pepper_row, pepper_col] = 0

    return noisy_img

# To read video from camera example:
NOISE_DENSITY = 0.005

SHARPEN_KERNEL = np.array([
    [ 0, -1,  0],
    [-1,  5, -1],
    [ 0, -1,  0]
])

MEDIAN_KERNEL = 5 

camera_id = 0
delay = 400
window_name = 'frame'

cap = cv.VideoCapture(camera_id)

cv.namedWindow('Original Color Feed', cv.WINDOW_NORMAL)
cv.namedWindow('Grayscale', cv.WINDOW_NORMAL)
cv.namedWindow('Noisy Grayscale', cv.WINDOW_NORMAL)
cv.namedWindow('Median Filtered', cv.WINDOW_NORMAL)
cv.namedWindow('Sharpened', cv.WINDOW_NORMAL)

if not cap.isOpened():
    sys.exit()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    cv.imshow('Original Color Feed', frame)

    gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    cv.imshow('Grayscale', gray_frame)

    noisy_frame = add_salt_pepper_noise(gray_frame, density=NOISE_DENSITY)
    cv.imshow('Noisy Grayscale', noisy_frame)

    denoised_frame = cv.medianBlur(noisy_frame, MEDIAN_KERNEL)
    cv.imshow('Median Filtered', denoised_frame)

    sharpened_frame = cv.filter2D(denoised_frame, -1, SHARPEN_KERNEL)
    cv.imshow('Sharpened', sharpened_frame)

    if cv.waitKey(delay) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()



```
